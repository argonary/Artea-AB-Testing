# Artea: A/B Testing and Targeted Coupon Strategy

## Business Context

Artea is an online retailer specializing in handmade clothing and accessories.
Despite healthy engagement metrics, 87% of website visitors never made a purchase.

To address this, Artea ran a controlled experiment: 5,000 recent visitors who had
not yet transacted were identified, and half were randomly assigned to receive a
**20% discount coupon** valid for one month. The remaining half received nothing.

This notebook analyzes the results of that experiment and builds a targeting model
to answer a harder question: given that coupons have a cost, **who should actually
receive one in the next campaign?**

## Analytical Roadmap

1. Data loading and validation
2. Exploratory data analysis
3. A/B test analysis -- did the coupon work?
4. Heterogeneous treatment effects -- who responded most?
5. Uplift modeling -- T-learner approach
6. Next campaign targeting -- scoring 6,000 new users
7. Business recommendations

## Section 1: Data Loading and Validation

Before any analysis, we load both datasets and confirm they match expectations:
the AB test data should have 5,000 rows with treatment and outcome columns, and
the Next Campaign data should have 6,000 rows with no outcome columns. We also
map the `channel_acq` codes to readable labels and cap `trans_after` at 1 to
treat it as a binary conversion outcome.

```python
# ── Load data ─────────────────────────────────────────────────────────────────

from pathlib import Path

DATA_PATH = Path("AB Data.xlsx")

ab   = pd.read_excel(DATA_PATH, sheet_name='AB_test')
next_camp = pd.read_excel(DATA_PATH, sheet_name='Next_Campaign')

# ── Channel mapping ───────────────────────────────────────────────────────────
channel_map = {1: 'Google', 2: 'Facebook', 3: 'Instagram', 4: 'Referral', 5: 'Other'}
ab['channel']        = ab['channel_acq'].map(channel_map)
next_camp['channel'] = next_camp['channel_acq'].map(channel_map)

# ── Binary conversion outcome ─────────────────────────────────────────────────
ab['converted'] = (ab['trans_after'] >= 1).astype(int)

# ── Validation ────────────────────────────────────────────────────────────────
print('AB Test shape:        ', ab.shape)
print('Next Campaign shape:  ', next_camp.shape)
print()
print('AB Test nulls:')
print(ab.isnull().sum())
print()
print('Next Campaign nulls:')
print(next_camp.isnull().sum())
print()
print('Treatment split:')
print(ab['test_coupon'].value_counts())
```

```python
# ── Preview ───────────────────────────────────────────────────────────────────
print('AB Test sample:')
display(ab.head())
print()
print('Data types:')
display(ab.dtypes)
print()
print('Basic statistics:')
display(ab.describe().round(2))
```

```python
# ── Randomization check ───────────────────────────────────────────────────────
# Verify treatment and control groups are balanced on pre-experiment features.
# A well-randomized experiment should show no statistically significant differences.

control   = ab[ab['test_coupon'] == 0]
treatment = ab[ab['test_coupon'] == 1]

check_cols = ['num_past_purch', 'spent_last_purchase',
              'weeks_since_visit', 'browsing_minutes', 'shopping_cart']

rows = []
for col in check_cols:
    t_stat, p_val = ttest_ind(control[col], treatment[col])
    rows.append({
        'Feature':          col,
        'Control Mean':     control[col].mean(),
        'Treatment Mean':   treatment[col].mean(),
        'Difference':       treatment[col].mean() - control[col].mean(),
        'p-value':          p_val,
        'Balanced (p>0.05)': 'Yes' if p_val > 0.05 else 'No'
    })

# Channel balance check
channel_crosstab = pd.crosstab(ab['channel'], ab['test_coupon'])
chi2, p_channel, _, _ = chi2_contingency(channel_crosstab)

rand_df = pd.DataFrame(rows).set_index('Feature')
display(rand_df.round(3))
print(f'\nChannel balance chi-square p-value: {p_channel:.3f}',
      '-- Balanced' if p_channel > 0.05 else '-- IMBALANCED')
```

### Randomization Check

Before analyzing outcomes, we verify that the random assignment produced balanced
groups. A t-test on each pre-experiment feature and a chi-square test on channel
distribution confirm no statistically significant differences between treatment
and control at the 5% significance level. The experiment is well-randomized and
any outcome differences can be attributed to the coupon rather than pre-existing
user characteristics.

## Section 3: A/B Test Analysis

With randomization confirmed, we can now measure the causal effect of the coupon.
We examine two outcome metrics:

- **Conversion rate**: did the coupon increase the probability of making a purchase?
- **Revenue per user**: did the coupon increase revenue, net of the discount?

We use a two-tailed z-test for conversion rates (proportions) and a Welch t-test
for revenue per user (continuous, unequal variances). All tests use a significance
level of 0.05.

```python
# ── Conversion rate test ───────────────────────────────────────────────────────
n_ctrl  = len(control)
n_treat = len(treatment)

conv_ctrl  = control['converted'].sum()
conv_treat = treatment['converted'].sum()

rate_ctrl  = conv_ctrl  / n_ctrl
rate_treat = conv_treat / n_treat

# Two-proportion z-test
p_pool = (conv_ctrl + conv_treat) / (n_ctrl + n_treat)
se     = np.sqrt(p_pool * (1 - p_pool) * (1/n_ctrl + 1/n_treat))
z_stat = (rate_treat - rate_ctrl) / se
p_val_conv = 2 * (1 - norm.cdf(abs(z_stat)))

# 95% confidence interval on the lift
se_diff = np.sqrt(rate_ctrl*(1-rate_ctrl)/n_ctrl + rate_treat*(1-rate_treat)/n_treat)
ci_low  = (rate_treat - rate_ctrl) - 1.96 * se_diff
ci_high = (rate_treat - rate_ctrl) + 1.96 * se_diff

print('── Conversion Rate Results ──────────────────────────────')
print(f'Control conversion rate:    {rate_ctrl:.1%}')
print(f'Treatment conversion rate:  {rate_treat:.1%}')
print(f'Absolute lift:              {rate_treat - rate_ctrl:.1%}')
print(f'Relative lift:              {(rate_treat - rate_ctrl) / rate_ctrl:.1%}')
print(f'95% CI on lift:             [{ci_low:.1%}, {ci_high:.1%}]')
print(f'Z-statistic:                {z_stat:.3f}')
print(f'P-value:                    {p_val_conv:.4f}')
print(f'Statistically significant:  {"Yes" if p_val_conv < 0.05 else "No"}')
```

### Conversion Rate Results

The coupon increased conversion from 10.7% to 12.4%, a 1.7 percentage point
absolute lift and 15.9% relative lift. However, this result falls just short of
conventional statistical significance (p = 0.060, two-tailed z-test). The 95%
confidence interval on the lift spans -0.1% to 3.5%, crossing zero at the lower
bound.

This marginal result does not mean the coupon was ineffective -- the direction
and magnitude of the effect are economically meaningful. It does suggest that
sending coupons indiscriminately to all users is unlikely to be the optimal
strategy, which motivates a more targeted approach.

```python
# ── Revenue per user test ─────────────────────────────────────────────────────
rev_ctrl  = control['revenue_after']
rev_treat = treatment['revenue_after']

t_stat, p_val_rev = ttest_ind(rev_treat, rev_ctrl, equal_var=False)

rev_lift    = rev_treat.mean() - rev_ctrl.mean()
se_rev      = np.sqrt(rev_treat.var()/n_treat + rev_ctrl.var()/n_ctrl)
ci_rev_low  = rev_lift - 1.96 * se_rev
ci_rev_high = rev_lift + 1.96 * se_rev

print('── Revenue Per User Results ─────────────────────────────')
print(f'Control avg revenue per user:    ${rev_ctrl.mean():.2f}')
print(f'Treatment avg revenue per user:  ${rev_treat.mean():.2f}')
print(f'Absolute lift:                   ${rev_lift:.2f}')
print(f'Relative lift:                   {rev_lift / rev_ctrl.mean():.1%}')
print(f'95% CI on lift:                  [${ci_rev_low:.2f}, ${ci_rev_high:.2f}]')
print(f'T-statistic:                     {t_stat:.3f}')
print(f'P-value:                         {p_val_rev:.4f}')
print(f'Statistically significant:       {"Yes" if p_val_rev < 0.05 else "No"}')
```

### Revenue Per User Results

Despite a modest lift in conversion rate, the coupon produced no statistically
significant increase in revenue per user. Treatment users generated an average
of USD 7.54 versus USD 7.78 for control users -- a difference of -USD 0.24
that is far from significant (p = 0.718).

This pattern is consistent with coupon cannibalization: a portion of treatment
users who converted would have purchased anyway, and the 20% discount simply
reduced the revenue collected from those transactions. Sent indiscriminately,
the coupon transfers margin to customers without generating incremental revenue.

This finding directly motivates a targeted approach: identify users whose
purchase probability is meaningfully higher with the coupon than without it,
and restrict distribution to that group.

```python
# ── A/B results summary chart ─────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(11, 5))

# Conversion rate comparison
ax = axes[0]
bars = ax.bar(['Control', 'Treatment'], [rate_ctrl * 100, rate_treat * 100],
              color=[BLUE, ORANGE], alpha=0.85, width=0.5)
for bar, val in zip(bars, [rate_ctrl * 100, rate_treat * 100]):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1,
            f'{val:.1f}%', ha='center', va='bottom', fontsize=11)
ax.set_title('Conversion Rate by Group')
ax.set_ylabel('Conversion Rate (%)')
ax.set_ylim(0, rate_treat * 100 * 1.3)
ax.text(0.5, 0.92, f'p = {p_val_conv:.3f}', transform=ax.transAxes,
        ha='center', fontsize=9, color=GRAY)

# Revenue per user comparison
ax = axes[1]
bars = ax.bar(['Control', 'Treatment'], [rev_ctrl.mean(), rev_treat.mean()],
              color=[BLUE, ORANGE], alpha=0.85, width=0.5)
for bar, val in zip(bars, [rev_ctrl.mean(), rev_treat.mean()]):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05,
            f'${val:.2f}', ha='center', va='bottom', fontsize=11)
ax.set_title('Avg Revenue Per User by Group')
ax.set_ylabel('Revenue Per User (USD)')
ax.set_ylim(0, rev_ctrl.mean() * 1.4)
ax.text(0.5, 0.92, f'p = {p_val_rev:.3f}', transform=ax.transAxes,
        ha='center', fontsize=9, color=GRAY)

plt.suptitle('A/B Test Results: Coupon Effect on Conversion and Revenue',
             fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('fig_ab_results.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Section 4: Heterogeneous Treatment Effects

The aggregate results mask important variation. Not all users respond to the coupon
equally -- some segments may show a strong positive response while others show none.
Identifying these segments is the foundation of the targeting model.

We examine coupon lift across acquisition channel, shopping cart behavior, and
past purchase history. For each subgroup we compute the conversion rate in
treatment and control separately, and the resulting lift.

```python
# ── Subgroup lift function with Bonferroni correction ─────────────────────────
def subgroup_lift(df, groupby_col):
    rows = []
    groups = df[groupby_col].unique()
    n_groups = len(groups)
    alpha_adj = 0.05 / n_groups

    for val, grp in df.groupby(groupby_col):
        ctrl  = grp[grp['test_coupon'] == 0]
        treat = grp[grp['test_coupon'] == 1]
        if len(ctrl) < 10 or len(treat) < 10:
            continue
        r_ctrl  = ctrl['converted'].mean()
        r_treat = treat['converted'].mean()
        lift    = r_treat - r_ctrl
        n_c, n_t = len(ctrl), len(treat)
        p_pool   = (ctrl['converted'].sum() + treat['converted'].sum()) / (n_c + n_t)
        se       = np.sqrt(p_pool * (1 - p_pool) * (1/n_c + 1/n_t))
        z        = lift / se if se > 0 else 0
        p        = 2 * (1 - norm.cdf(abs(z)))
        rows.append({
            groupby_col:            val,
            'N Control':            n_c,
            'N Treatment':          n_t,
            'Control Conv %':       round(r_ctrl * 100, 1),
            'Treatment Conv %':     round(r_treat * 100, 1),
            'Lift (pp)':            round(lift * 100, 1),
            'p-value':              round(p, 3),
            'Sig (Bonferroni)':     'Yes' if p < alpha_adj else 'No'
        })
    result = pd.DataFrame(rows).set_index(groupby_col)
    print(f'Bonferroni-adjusted threshold ({n_groups} groups): p < {alpha_adj:.4f}')
    return result

channel_lift = subgroup_lift(ab, 'channel')
display(channel_lift)
```

### Subgroup Lift by Acquisition Channel

After applying a Bonferroni correction for five simultaneous tests (adjusted
threshold: p < 0.010), no individual channel reaches statistical significance.
However, the directional pattern is informative.

Facebook users show the largest lift (5.0 percentage points, p = 0.025), suggesting
stronger coupon responsiveness among this segment. Google users -- the largest
channel at 40.7% of the sample -- show a slightly negative lift (-1.4 pp),
indicating the coupon may be largely wasted on this group.

These patterns motivate the uplift model, which identifies responsive users at
the individual level rather than relying on subgroup averages.

```python
# ── Subgroup lift chart by channel ────────────────────────────────────────────
channel_order = channel_lift['Lift (pp)'].sort_values(ascending=True).index

fig, ax = plt.subplots(figsize=(8, 5))

colors = [ORANGE if v >= 0 else '#C0392B' for v in channel_lift.loc[channel_order, 'Lift (pp)']]
bars = ax.barh(channel_order, channel_lift.loc[channel_order, 'Lift (pp)'],
               color=colors, alpha=0.85, height=0.5)

for bar, val, p in zip(bars,
                        channel_lift.loc[channel_order, 'Lift (pp)'],
                        channel_lift.loc[channel_order, 'p-value']):
    x_pos = val + 0.1 if val >= 0 else val - 0.1
    ha    = 'left' if val >= 0 else 'right'
    ax.text(x_pos, bar.get_y() + bar.get_height() / 2,
            f'{val:+.1f} pp  (p={p:.3f})', va='center', ha=ha, fontsize=9)

ax.axvline(0, color=GRAY, linewidth=1.0, linestyle='--')
ax.set_xlabel('Conversion Rate Lift (percentage points)')
ax.set_title('Coupon Lift by Acquisition Channel\n(Bonferroni-adjusted threshold: p < 0.010)')
ax.set_xlim(channel_lift['Lift (pp)'].min() - 4, channel_lift['Lift (pp)'].max() + 6)

plt.tight_layout()
plt.savefig('fig_channel_lift.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Subgroup lift: shopping cart and past purchases ───────────────────────────
cart_lift = subgroup_lift(ab, 'shopping_cart')
cart_lift.index = cart_lift.index.map({0: 'No Cart', 1: 'Cart Item'})
print('\nShopping Cart Subgroups:')
display(cart_lift)

# Bin past purchases into meaningful groups
ab['purch_bin'] = pd.cut(ab['num_past_purch'],
                          bins=[-1, 0, 2, 5, 100],
                          labels=['0 purchases', '1-2 purchases',
                                  '3-5 purchases', '6+ purchases'])

purch_lift = subgroup_lift(ab, 'purch_bin')
print('\nPast Purchase Subgroups:')
display(purch_lift)
```

### Subgroup Lift: Shopping Cart and Past Purchases

**Shopping cart behavior** is a strong moderator of coupon response. Users who
added an item to their cart without purchasing show a 4.6 percentage point lift
(p = 0.028), approaching the Bonferroni-adjusted threshold. These high-intent
users appear to need only a small incentive to complete a purchase. Users with
no cart activity show negligible lift (0.7 pp), suggesting the coupon alone is
insufficient to create purchase intent from scratch.

**Past purchase history** reveals a non-linear pattern. Early repeat customers
(1-2 prior purchases) show the strongest lift among returning users (3.0 pp,
p = 0.024). Heavy buyers (6+ purchases) show a slightly negative lift, consistent
with the Sure Thing problem: frequent buyers purchase regardless of the coupon,
so the discount simply erodes margin. New customers (0 purchases) show modest
but non-significant lift, suggesting acquisition requires more than a price nudge.

```python
# ── Subgroup lift charts: cart and past purchases ─────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Shopping cart
ax = axes[0]
colors = [ORANGE if v >= 0 else '#C0392B' for v in cart_lift['Lift (pp)']]
bars = ax.barh(cart_lift.index, cart_lift['Lift (pp)'],
               color=colors, alpha=0.85, height=0.4)
for bar, val, p in zip(bars, cart_lift['Lift (pp)'], cart_lift['p-value']):
    x_pos = val + 0.1 if val >= 0 else val - 0.1
    ha    = 'left' if val >= 0 else 'right'
    ax.text(x_pos, bar.get_y() + bar.get_height() / 2,
            f'{val:+.1f} pp  (p={p:.3f})', va='center', ha=ha, fontsize=9)
ax.axvline(0, color=GRAY, linewidth=1.0, linestyle='--')
ax.set_xlabel('Conversion Rate Lift (percentage points)')
ax.set_title('Coupon Lift by Shopping Cart\n(Bonferroni threshold: p < 0.025)')
ax.set_xlim(-2, cart_lift['Lift (pp)'].max() + 5)

# Past purchases
ax = axes[1]
purch_order = ['0 purchases', '1-2 purchases', '3-5 purchases', '6+ purchases']
purch_vals  = purch_lift.loc[purch_order, 'Lift (pp)']
purch_pvals = purch_lift.loc[purch_order, 'p-value']
colors = [ORANGE if v >= 0 else '#C0392B' for v in purch_vals]
bars = ax.barh(purch_order, purch_vals, color=colors, alpha=0.85, height=0.4)
for bar, val, p in zip(bars, purch_vals, purch_pvals):
    x_pos = val + 0.15 if val >= 0 else val - 0.15
    ha    = 'left' if val >= 0 else 'right'
    ax.text(x_pos, bar.get_y() + bar.get_height() / 2,
            f'{val:+.1f} pp  (p={p:.3f})', va='center', ha=ha, fontsize=9)
ax.axvline(0, color=GRAY, linewidth=1.0, linestyle='--')
ax.set_xlabel('Conversion Rate Lift (percentage points)')
ax.set_title('Coupon Lift by Past Purchases\n(Bonferroni threshold: p < 0.013)')
ax.set_xlim(-4, purch_vals.max() + 6)
ax.set_yticklabels(purch_order, fontsize=10)

plt.suptitle('Heterogeneous Treatment Effects', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('fig_subgroup_lift.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Section 5: Uplift Modeling -- T-Learner Approach

The aggregate A/B test showed a marginal overall effect. The subgroup analysis
revealed that coupon responsiveness varies substantially across users. The next
step is to move from group-level averages to individual-level predictions.

We use the T-learner, a meta-learner approach from the causal ML literature
(Kunzel et al., 2019). Two logistic regression models are trained separately:

- **Model T**: trained on the treatment group only. Learns the relationship
  between user features and purchase probability when a coupon is received.
- **Model C**: trained on the control group only. Learns the relationship
  between user features and purchase probability without a coupon.

For any user, the uplift score is:

    Uplift = P(purchase | coupon) - P(purchase | no coupon)

Users with high uplift scores are **Persuadables** -- they buy with a coupon
but not without one. Users with low Model C scores and low uplift are **Lost
Causes**. Users with high Model C scores regardless of treatment are **Sure
Things** -- giving them a coupon wastes margin.

```python
# ── Feature setup ─────────────────────────────────────────────────────────────
FEATURES = ['num_past_purch', 'spent_last_purchase', 'weeks_since_visit',
            'browsing_minutes', 'shopping_cart', 'channel_acq']

TARGET = 'converted'

treat_df = ab[ab['test_coupon'] == 1].copy()
ctrl_df  = ab[ab['test_coupon'] == 0].copy()

X_treat = treat_df[FEATURES]
y_treat = treat_df[TARGET]

X_ctrl  = ctrl_df[FEATURES]
y_ctrl  = ctrl_df[TARGET]

X_all   = ab[FEATURES]

# ── Train Model T and Model C ─────────────────────────────────────────────────
pipe_T = Pipeline([
    ('scaler', StandardScaler()),
    ('model',  LogisticRegression(max_iter=1000, random_state=42, C=1.0))
])

pipe_C = Pipeline([
    ('scaler', StandardScaler()),
    ('model',  LogisticRegression(max_iter=1000, random_state=42, C=1.0))
])

pipe_T.fit(X_treat, y_treat)
pipe_C.fit(X_ctrl,  y_ctrl)

# ── Uplift scores on full AB dataset ─────────────────────────────────────────
ab['p_treat'] = pipe_T.predict_proba(X_all)[:, 1]
ab['p_ctrl']  = pipe_C.predict_proba(X_all)[:, 1]
ab['uplift']  = ab['p_treat'] - ab['p_ctrl']

print('Model T -- trained on treatment group')
print(f'  Observations: {len(X_treat)}')
print(f'  Conversion rate: {y_treat.mean():.1%}')
print()
print('Model C -- trained on control group')
print(f'  Observations: {len(X_ctrl)}')
print(f'  Conversion rate: {y_ctrl.mean():.1%}')
print()
print('Uplift score summary (full AB dataset):')
print(ab['uplift'].describe().round(4))
```

```python
# ── Cross-validated AUC for Model T and Model C ───────────────────────────────
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

auc_T = roc_auc_score(y_treat,
          cross_val_predict(pipe_T, X_treat, y_treat, cv=cv, method='predict_proba')[:, 1])

auc_C = roc_auc_score(y_ctrl,
          cross_val_predict(pipe_C, X_ctrl, y_ctrl, cv=cv, method='predict_proba')[:, 1])

print('Cross-validated AUC scores (5-fold):')
print(f'  Model T (treatment group): {auc_T:.3f}')
print(f'  Model C (control group):   {auc_C:.3f}')
print()
print('Note: AUC > 0.5 confirms models predict better than random.')
print('Low AUC is expected given limited features -- the uplift')
print('score is still informative as a ranking signal.')
```

### Model Evaluation

Both models achieve strong discriminative performance on cross-validated AUC:

- Model T (treatment group): AUC = 0.799
- Model C (control group): AUC = 0.807

An AUC of 0.80 means the model correctly ranks a randomly chosen converter above
a randomly chosen non-converter 80% of the time. This is a solid result given
that we have only six input features. Both models are well-calibrated relative
to each other, which is important for the uplift score to be meaningful -- if
one model were systematically better than the other, the difference score would
be biased.

- **Sure Things (36.8%)**: High baseline probability and positive uplift. These

The optimal targeting strategy focuses on Persuadables while avoiding Sure Things,

## Section 6: Next Campaign Targeting

With the T-learner trained on the A/B test data, we now score the 6,000 users
in the Next Campaign dataset. These users have no observed outcome -- they were
not part of the experiment. The uplift model assigns each user a predicted
incremental conversion probability if they were to receive a coupon.

We then simulate three targeting strategies and compare their expected performance:

- **Target nobody**: no coupons sent, no incremental conversions, no coupon cost
- **Target everybody**: 6,000 coupons sent, high reach but significant waste
- **Target by uplift**: send coupons only to users above an uplift threshold,
  maximizing incremental conversions per coupon sent

```python
# ── Score Next Campaign users ─────────────────────────────────────────────────
X_next = next_camp[FEATURES]

next_camp['p_treat'] = pipe_T.predict_proba(X_next)[:, 1]
next_camp['p_ctrl']  = pipe_C.predict_proba(X_next)[:, 1]
next_camp['uplift']  = next_camp['p_treat'] - next_camp['p_ctrl']

print('Next Campaign uplift score summary:')
print(next_camp['uplift'].describe().round(4))
print()
print(f'Users with positive uplift: {(next_camp["uplift"] > 0).sum():,} '
      f'({(next_camp["uplift"] > 0).mean():.1%})')
print(f'Users with uplift > 0.05:   {(next_camp["uplift"] > 0.05).sum():,} '
      f'({(next_camp["uplift"] > 0.05).mean():.1%})')
print(f'Users with uplift > 0.10:   {(next_camp["uplift"] > 0.10).sum():,} '
      f'({(next_camp["uplift"] > 0.10).mean():.1%})')
```

```python
# ── Targeting threshold simulation ────────────────────────────────────────────
# Assumptions
COUPON_DISCOUNT = 0.20
AVG_ORDER_VALUE = avg_rev_conv / (1 - COUPON_DISCOUNT)  # gross AOV before discount
COUPON_COST_PER_CONVERSION = AVG_ORDER_VALUE * COUPON_DISCOUNT

N_NEXT          = len(next_camp)

thresholds = np.linspace(next_camp['uplift'].min(), next_camp['uplift'].max(), 300)

results = []
for t in thresholds:
    targeted        = next_camp[next_camp['uplift'] >= t]
    n_targeted      = len(targeted)
    if n_targeted == 0:
        continue
    # Expected incremental conversions from uplift scores
    inc_conversions = targeted['uplift'].sum()
    # Expected revenue: incremental converters pay net-of-discount price
    inc_revenue     = inc_conversions * AVG_ORDER_VALUE * (1 - COUPON_DISCOUNT)
    # Coupon cost is only the discount on incremental purchases
    # (we only count the true cost: margin given away on persuaded purchases)
    pct_targeted    = n_targeted / N_NEXT

    results.append({
        'threshold':      t,
        'n_targeted':     n_targeted,
        'pct_targeted':   pct_targeted,
        'inc_conversions': inc_conversions,
        'inc_revenue':    inc_revenue,
    })

sim_df = pd.DataFrame(results)

# ── Plot targeting curve ───────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Incremental conversions vs % targeted
ax = axes[0]
ax.plot(sim_df['pct_targeted'] * 100, sim_df['inc_conversions'],
        color=BLUE, linewidth=2)
ax.axvline(16.1, color=ORANGE, linestyle='--', linewidth=1.2, label='Uplift > 0.05 (16.1%)')
ax.axvline(5.5,  color=GRAY,   linestyle='--', linewidth=1.2, label='Uplift > 0.10 (5.5%)')
ax.set_xlabel('Percentage of Users Targeted (%)')
ax.set_ylabel('Expected Incremental Conversions')
ax.set_title('Incremental Conversions vs Targeting Rate')
ax.legend(fontsize=9)

# Incremental revenue vs % targeted
ax = axes[1]
ax.plot(sim_df['pct_targeted'] * 100, sim_df['inc_revenue'],
        color=ORANGE, linewidth=2)
ax.axvline(16.1, color=BLUE, linestyle='--', linewidth=1.2, label='Uplift > 0.05 (16.1%)')
ax.axvline(5.5,  color=GRAY,  linestyle='--', linewidth=1.2, label='Uplift > 0.10 (5.5%)')
ax.set_xlabel('Percentage of Users Targeted (%)')
ax.set_ylabel('Expected Incremental Revenue (USD)')
ax.set_title('Incremental Revenue vs Targeting Rate')
ax.legend(fontsize=9)

plt.suptitle('Next Campaign: Targeting Threshold Analysis', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('fig_targeting_curve.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Three strategy comparison ─────────────────────────────────────────────────
def strategy_summary(label, targeted_df, all_df):
    n_targeted     = len(targeted_df)
    inc_conv       = targeted_df['uplift'].sum()
    inc_revenue    = inc_conv * AVG_ORDER_VALUE * (1 - COUPON_DISCOUNT)
    pct_targeted   = n_targeted / len(all_df)
    return {
        'Strategy':              label,
        'Coupons Sent':          n_targeted,
        'Pct Targeted':          f'{pct_targeted:.1%}',
        'Incremental Conversions': round(inc_conv, 1),
        'Incremental Revenue':   f'${inc_revenue:,.0f}',
    }

nobody    = next_camp[next_camp['uplift'] > 999]   # empty
everybody = next_camp
top_16    = next_camp[next_camp['uplift'] >= 0.05]
top_5     = next_camp[next_camp['uplift'] >= 0.10]

strategies = [
    strategy_summary('Target nobody',        nobody,    next_camp),
    strategy_summary('Target everybody',     everybody, next_camp),
    strategy_summary('Target uplift > 0.05', top_16,   next_camp),
    strategy_summary('Target uplift > 0.10', top_5,    next_camp),
]

strat_df = pd.DataFrame(strategies).set_index('Strategy')
display(strat_df)
```

```python
# ── Cost-adjusted strategy comparison ────────────────────────────────────────
rows = []
for label, targeted_df in [
    ('Target nobody',        nobody),
    ('Target everybody',     everybody),
    ('Target uplift > 0.05', top_16),
    ('Target uplift > 0.10', top_5),
]:
    n_targeted   = len(targeted_df)
    inc_conv     = targeted_df['uplift'].sum()
    inc_revenue  = inc_conv * AVG_ORDER_VALUE * (1 - COUPON_DISCOUNT)
    coupon_cost  = inc_conv * COUPON_COST_PER_CONVERSION
    net_profit   = inc_revenue - coupon_cost
    coupons_per_conv = n_targeted / inc_conv if inc_conv > 0 else 0
    profit_per_coupon = net_profit / n_targeted if n_targeted > 0 else 0

    rows.append({
        'Strategy':               label,
        'Coupons Sent':           n_targeted,
        'Inc. Conversions':       round(inc_conv, 1),
        'Net Incremental Profit': f'${net_profit:,.0f}',
        'Coupons per Conversion': round(coupons_per_conv, 1),
        'Profit per Coupon Sent': f'${profit_per_coupon:.2f}',
    })

cost_df = pd.DataFrame(rows).set_index('Strategy')
display(cost_df)

print()
print(f'Assumed average order value (gross): ${AVG_ORDER_VALUE:.2f}')
print(f'Coupon discount:                      20%')
print(f'Coupon cost per redemption:          ${COUPON_COST_PER_CONVERSION:.2f}')
```

### Targeting Strategy Comparison

Scoring the 6,000 Next Campaign users with the T-learner and simulating targeting
thresholds reveals a clear efficiency frontier.

A blanket campaign sending 6,000 coupons generates an estimated 121 incremental
conversions but requires 49.6 coupons per conversion at USD 1.00 net profit per
coupon sent -- barely above break-even on the targeting decision itself.

Restricting coupons to users with uplift above 0.05 (16.1% of the pool, 964 users)
captures 89 incremental conversions -- 73.6% of the blanket campaign total -- at
only 10.8 coupons per conversion and USD 4.60 net profit per coupon sent. This
represents a 4.6x improvement in targeting efficiency.

The uplift > 0.10 threshold (5.5% of users, 330 coupons) is the most efficient
at USD 6.60 per coupon sent and is appropriate under a tight budget constraint,
though it leaves more than half of available incremental conversions unreached.

The recommended strategy is to target users with uplift above 0.05, adjusting
the threshold downward or upward based on available coupon budget.

```python
# ── Strategy comparison chart ─────────────────────────────────────────────────
strategies_plot = ['Target everybody', 'Target uplift > 0.05', 'Target uplift > 0.10']
coupons_sent    = [6000, 964, 330]
inc_convs       = [121.0, 89.1, 43.8]
profit_per_cpn  = [1.00, 4.60, 6.60]
net_profits     = [6026, 4434, 2179]

fig, axes = plt.subplots(1, 3, figsize=(13, 5))

# Coupons sent
ax = axes[0]
bars = ax.bar(strategies_plot, coupons_sent,
              color=[GRAY, ORANGE, BLUE], alpha=0.85, width=0.5)
for bar, val in zip(bars, coupons_sent):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
            f'{val:,}', ha='center', va='bottom', fontsize=10)
ax.set_title('Coupons Sent')
ax.set_ylabel('Number of Coupons')
ax.set_xticklabels(strategies_plot, fontsize=8.5)
ax.set_ylim(0, max(coupons_sent) * 1.2)

# Net incremental profit
ax = axes[1]
bars = ax.bar(strategies_plot, net_profits,
              color=[GRAY, ORANGE, BLUE], alpha=0.85, width=0.5)
for bar, val in zip(bars, net_profits):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
            f'${val:,}', ha='center', va='bottom', fontsize=10)
ax.set_title('Net Incremental Profit')
ax.set_ylabel('USD')
ax.set_xticklabels(strategies_plot, fontsize=8.5)
ax.set_ylim(0, max(net_profits) * 1.2)

# Profit per coupon sent
ax = axes[2]
bars = ax.bar(strategies_plot, profit_per_cpn,
              color=[GRAY, ORANGE, BLUE], alpha=0.85, width=0.5)
for bar, val in zip(bars, profit_per_cpn):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05,
            f'${val:.2f}', ha='center', va='bottom', fontsize=10)
ax.set_title('Net Profit per Coupon Sent')
ax.set_ylabel('USD per Coupon')
ax.set_xticklabels(strategies_plot, fontsize=8.5)
ax.set_ylim(0, max(profit_per_cpn) * 1.3)

plt.suptitle('Targeting Strategy Comparison', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('fig_strategy_comparison.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Section 7: Business Recommendations

This section synthesizes findings from the A/B test, heterogeneous treatment
effect analysis, and uplift model into actionable recommendations for Artea's
coupon strategy.

### Key Findings

**1. The coupon produced a marginal but directionally positive overall effect.**
The 20% discount coupon increased conversion from 10.7% to 12.4% (1.7 percentage
point absolute lift, p = 0.060). This result falls just short of conventional
statistical significance, suggesting that indiscriminate coupon distribution is
unlikely to be a reliably profitable strategy at scale.

**2. Revenue per user did not increase under a blanket campaign.**
Despite the modest conversion lift, average revenue per user was essentially
unchanged between treatment (USD 7.54) and control (USD 7.78), p = 0.718.
The 20% discount eroded margins on purchases that would have occurred regardless,
confirming CEO Alex Campbel's concern about coupon cannibalization.

**3. Coupon responsiveness is concentrated in specific user segments.**
Subgroup analysis identified two high-response profiles:
- Users who added items to their shopping cart without purchasing (4.6 pp lift)
- Users acquired via Facebook (5.0 pp lift)
Both findings approach but do not cross the Bonferroni-adjusted significance
threshold, warranting caution but informing the targeting model directionally.

**4. Heavy prior purchasers should not receive coupons.**
Users with 6 or more prior purchases showed a slightly negative coupon lift,
consistent with the Sure Thing problem. These customers purchase organically
and discounting their transactions destroys margin without generating incremental
revenue.

**5. Targeted distribution dramatically outperforms blanket campaigns.**
The T-learner uplift model identifies a subset of 964 users (16.1% of the next
campaign pool) with predicted uplift above 0.05. Targeting this group captures
73.6% of available incremental conversions at 4.6x the profit efficiency of a
blanket campaign (USD 4.60 vs USD 1.00 net profit per coupon sent).

### Recommended Strategy

Target the next campaign to users with a T-learner uplift score above 0.05.
This threshold balances incremental conversion volume against coupon efficiency.
Under a tighter budget constraint, raise the threshold to 0.10, which targets
330 users at USD 6.60 net profit per coupon sent.

Prioritize users with the following profile:
- Added an item to their shopping cart in their last visit
- Acquired via Facebook or Instagram
- Have 1-2 prior purchases (early repeat customers, not yet habitual buyers)
- Have not visited in 2-4 weeks (recent enough to remember Artea, distant
  enough that a nudge adds value)

### Limitations and Next Steps

- The uplift model is a T-learner using logistic regression -- a strong baseline
  but not the most powerful method available. A follow-on analysis could implement
  an X-learner or causal forest to capture non-linear interactions.
- Subgroup findings (Facebook, shopping cart) did not reach Bonferroni-adjusted
  significance. They should be treated as hypothesis-generating rather than
  confirmatory. A follow-up experiment targeting these segments specifically
  would provide cleaner evidence.
- The revenue model assumes the coupon cost is limited to discount on incremental
  conversions. In practice, Sure Things who receive a coupon and redeem it also
  represent a margin cost not captured here.
- The next campaign dataset has no observed outcomes. Model performance on this
  population cannot be validated without running another experiment.
